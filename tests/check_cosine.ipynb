```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.optim.lr_scheduler import CosineAnnealingLR
```

```python
l = nn.Linear(5, 1)
o = optim.SGD(l.parameters(), 1e-1)
s = CosineAnnealingLR(o, T_max=200)
```

```python
lrs = []
for i in range(200):
    s.step()
    lrs.append(s.get_lr())
```

```python
import matplotlib.pyplot as plt
plt.plot(lrs)
```
